# Quiz: Hot Spot Scenarios in System Design

## Which of the following scenarios could realistically introduce a hot spot in a system?

---

### ✅ 1. The system gets a sudden surge of traffic from a particular region in the world

**✔️ This can cause a hot spot.**

**Explanation:**  
A hot spot occurs when a particular part of the system gets significantly more load than the rest. A sudden surge of traffic from a specific region can cause:

- Regional servers to overload
- Cache warmups or invalidations for that region
- CDN edge nodes or databases tied to that region to spike in activity

**Example:**  
If a ticketing site for a cricket tournament goes live, and millions of users from India try to access it simultaneously, only the servers handling Indian traffic might become hot spots due to uneven load.

---

### ❌ 2. A system's load balancer uses a round robin server selection strategy

**❌ This generally doesn't cause a hot spot (on its own).**

**Explanation:**  
A **round robin** load balancing strategy distributes incoming requests evenly across all backend servers in a rotating fashion. It’s **designed** to **prevent hot spots**.

However, **if the backend servers differ in capacity**, then round robin may unintentionally overload weaker servers — but **round robin itself** is **not a direct cause** of a hot spot.

**Example (not a hot spot):**  
If there are 5 servers and 100 requests, each server gets 20 requests — evenly distributed.

---

### ✅ 3. The asynchronous replication of a system's primary database fails

**✔️ This can cause a hot spot.**

**Explanation:**  
If the **asynchronous replica fails**, all read or write traffic may be **rerouted to the primary** database. This puts sudden extra pressure on one node (the primary), creating a **hot spot at the database layer**.

**Example:**  
- You have a primary DB and 2 replicas used for reading.  
- If the replicas go down, all reads hit the primary.  
- The primary DB gets overwhelmed, slowing down the entire system — classic hot spot.

---

### ✅ 4. A new social media trend encourages everyone named Alex or Alexa to tweet hello

**✔️ This can cause a hot spot — depending on how your system stores or indexes data.**

**Explanation:**  
If your system **indexes or shards** user data by **username or first name**, and suddenly **millions of users named Alex or Alexa** perform the same action, then:

- All related records or database partitions for “Alex/Alexa” will be overwhelmed.
- You’ll see a **skewed access pattern** — leading to hot spots.

**Example:**  
Let’s say you hash usernames to assign them to database partitions:  
- All usernames starting with “A” go to Partition 1  
- A massive number of Alex/Alexa users tweeting at once → Partition 1 spikes → hot spot

---

## ✅ Final Answer Summary:

| Scenario                                                                 | Is Hot Spot Possible? | Why                                                                 |
|-------------------------------------------------------------------------|------------------------|----------------------------------------------------------------------|
| Sudden surge of traffic from a region                                   | ✅ Yes                | Overloads specific regional servers                                 |
| Round robin load balancer strategy                                      | ❌ No                 | Distributes traffic evenly unless backend nodes are uneven          |
| Failure of async DB replication                                         | ✅ Yes                | All traffic hits primary DB                                         |
| Social trend from users named Alex/Alexa                                | ✅ Yes                | Causes skewed access to specific shards/partitions                  |
